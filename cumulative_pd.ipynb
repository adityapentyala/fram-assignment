```python
import pandas as pd
```

```python
pd_df = pd.read_excel("outputs/probability_of_default_year_2_to_10.xlsx", index_col=0)
```

```python
probability_survival_df = 1-pd_df
probability_survival_df
```

```python
cumulative_pd_df = pd.DataFrame(1, index=pd_df.index, columns=['Year 5', 'Year 10'])
prod = pd.DataFrame(1, index=pd_df.index, columns=['dummy'])
for year in range(2, 11):
    prod['dummy'] = prod['dummy'] * probability_survival_df[f"Year {year}"]
    if year == 5 or year == 10:
        cumulative_pd_df[f'Year {year}'] = 1 - prod['dummy']
cumulative_pd_df
```

```python
cumulative_pd_df.to_excel("outputs/cumulative_probability_of_default_year_5_and_10.xlsx", index_label="Rating")
```
