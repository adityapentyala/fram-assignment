```python
import pandas as pd
```

```python
ratings_df = pd.read_excel('data/ratings_data.xlsx')
ratings_df
```

```python
ratings = ['AAA', 'AA', 'A', 'BBB', 'BB', 'B', 'C', 'D']

transition_matrix = pd.DataFrame(0.0, index=ratings, columns=ratings)
transition_matrix
```

```python
for i in range(len(ratings_df)):
    for j in range(2, len(ratings_df.columns)):
        from_rating = ratings_df.iloc[i, j-1]
        to_rating = ratings_df.iloc[i, j]
        transition_matrix.loc[from_rating, to_rating] += 1
```

```python
transition_matrix
```

```python
for i in range(len(transition_matrix)):
    total = transition_matrix.iloc[i].sum()
    for j in range(len(transition_matrix.columns)):
        transition_matrix.iloc[i, j] /= total
transition_matrix
```

```python
transition_matrix.to_excel('outputs/transition_matrix.xlsx')
```
